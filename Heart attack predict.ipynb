```python
result = df.copy()
max_value = df['trtbps'].max()
min_value = df['trtbps'].min()
result['trtbps'] = (df['trtbps'] - min_value) / (max_value - min_value)

max_value = df['age'].max()
min_value = df['age'].min()
result['age'] = (df['age'] - min_value) / (max_value - min_value)

max_value = df['cp'].max()
min_value = df['cp'].min()
result['cp'] = (df['cp'] - min_value) / (max_value - min_value)

max_value = df['chol'].max()
min_value = df['chol'].min()
result['chol'] = (df['chol'] - min_value) / (max_value - min_value)

max_value = df['thalachh'].max()
min_value = df['thalachh'].min()
result['thalachh'] = (df['thalachh'] - min_value) / (max_value - min_value)

max_value = df['oldpeak'].max()
min_value = df['oldpeak'].min()
result['oldpeak'] = (df['oldpeak'] - min_value) / (max_value - min_value)

max_value = df['caa'].max()
min_value = df['caa'].min()
result['caa'] = (df['caa'] - min_value) / (max_value - min_value)

max_value = df['thall'].max()
min_value = df['thall'].min()
result['thall'] = (df['thall'] - min_value) / (max_value - min_value)
```

```python
from sklearn.model_selection import train_test_split
features=["age", "sex", "cp","trtbps","restecg", "thalachh","exng","oldpeak", "slp","thall"]
X=df[features]
y=df.output
train_X, test_X, train_y, test_y = train_test_split(X, y,train_size=0.8, test_size=0.2,random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score

model1=DecisionTreeRegressor(random_state=1)
model1.fit(train_X, train_y)
pred1=model1.predict(test_X)
print(metrics.accuracy_score(test_y, pred1))
print("Mean Absolute Error: %f" %(mean_absolute_error(test_y, pred1)))
```
